# Webscraping using beautiful soup

I've written a seperate program in python named (Web_Scraper_Using_Beautiful_Soup.py) which is found inside the repo  using the pycharm IDE since I had problems in implementing the beautiful soup package in my jupyter notebook. The web scraper generated a file called postal_codes csv which was scraped from https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
```

### importing the csv file generated by the web scraper

```python
postal_codes = pd.read_csv('postal_codes.csv')
postal_codes.head()
```

### stripping some unwanted characters

```python
postal_codes['Neighbourhood'] = postal_codes['Neighbourhood'].map(lambda x: x.rstrip('\r\n'))
```

```python
postal_codes.head()
```

### excluding rows with unassigned data in Borough column

```python
postal_codes = postal_codes[postal_codes.Borough != 'Not assigned']
postal_codes.head()
```

### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. 

```python
df = postal_codes.groupby('PostalCode', as_index=False).agg(lambda x: ', '.join(set(x)))
df.head(12)
```
